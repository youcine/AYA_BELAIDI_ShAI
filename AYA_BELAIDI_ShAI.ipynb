# 1. Basic Data Exploration

**importing Data and librearies**

```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv('/content/drive/MyDrive/Salaries.csv')
df.head()
```

```python
# Identify the number of rows and columns in the dataset and determine the data types of each column
df.info()
```

```python
# Check for missing values in each column
df.isnull().sum()
```

# 2. Descriptive Statistics

```python
#Calculate basic statistics mean, median, mode, minimum, and maximum salary and the standard deviation of salaries
df.describe()
```

```python
# Determine the range of salaries

salary_range = (df['TotalPay'].max() - df['TotalPay'].min())
print("Range of salaries:", salary_range)
```

# 3. Data Cleaning

```python
# Drop 'Notes' and 'Status' columns because it's all NAN
df= df.drop(['Notes', 'Status'], axis=1)

# Fill missing values in 'BasePay' and 'Benefits' with their respective means because it is suitable for the situation
#if i have more time and resources for this task i would calculate the mean for each job title because base pay and benefits
#differs from a job title to another and depens on the experience and years of work alse
df['BasePay'].fillna(df['BasePay'].mean(), inplace=True)
df['Benefits'].fillna(df['Benefits'].mean(), inplace=True)
df['OtherPay'].fillna(df['OtherPay'].mean(), inplace=True)
df['OvertimePay'].fillna(df['OvertimePay'].mean(), inplace=True)
```

```python
df.info()
```

# 4.Basic Data Visualization

```python
import matplotlib.pyplot as plt

# histogram to visualize the distribution of salaries
plt.figure(figsize=(12, 6))
plt.hist(df['TotalPay'], bins=50, edgecolor='gray', color='pink')

plt.title('Distribution of Salaries')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
# Use pie charts to represent the proportion of employees in different departments
department_counts = df['JobTitle'].value_counts().head(10)
# I only used 10 beacuse there is lot of job titels
sns.set_palette("pastel")
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Employees in Top 10 Departments')
plt.show()
```

# 5.Grouped Analysis

```python
# Step 1: Group by 'BasePay' and calculate summary statistics for 'OvertimePay', 'OtherPay', and 'Benefits'
grouped_data_basepay = df.groupby('BasePay').agg({
    'OvertimePay': 'mean',
    'OtherPay': 'mean',
    'Benefits': 'mean'
}).reset_index()
print("Summary Statistics by BasePay:")
print(grouped_data_basepay)

# Step 2: Compare the result with 'TotalPay'
# Group by 'BasePay' and calculate the mean of 'TotalPay'
grouped_data_totalpay = df.groupby('BasePay')['TotalPay'].mean().reset_index()

print("\nSummary Statistics for TotalPay:")
print(grouped_data_totalpay)
```

# 6. Simple Correlation Analysis

```python
# Identify correlation between 'TotalPay' and 'BasePay'
correlation = df['TotalPay'].corr(df['BasePay'])
print("Correlation between TotalPay and BasePay:", correlation)
```

```python
# Plot a scatter plot to visualize the relationship
plt.scatter(df['BasePay'], df['TotalPay'])
plt.title('Scatter Plot of TotalPay vs BasePay')
plt.xlabel('BasePay')
plt.ylabel('TotalPay')
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
