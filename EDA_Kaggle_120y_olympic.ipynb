```python
# the top 4 countries and years 1986 to 2016
top_countries = ['USA', 'FRA', 'GER', 'GBR']
years = range(1986, 2017)
subset_df = df[df['NOC'].isin(top_countries) & df['Year'].isin(years)]
plt.figure(figsize=(10, 6))
plt.title('Medals over Time for Top 4 Countries (1986-2016)', fontdict={'fontweight': 'bold', 'fontsize': 18})

for country in top_countries:
    country_medals = subset_df[subset_df['NOC'] == country].groupby('Year')['Medal'].count()
    plt.plot(country_medals.index, country_medals.values, marker='.', label=country)


plt.xlabel('Year')
plt.ylabel('Number of Medals')
plt.xticks(range(1986, 2017, 2))
plt.legend()

plt.show()
```
